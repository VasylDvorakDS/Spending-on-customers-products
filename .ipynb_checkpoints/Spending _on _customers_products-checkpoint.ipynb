## DBSCAN

**Поскольку значения разных признаков имеют разные масштабы единиц измерения, выполним масштабирование данных с помощью StandardScaler.**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
scaled_data
```

**С помощью DBSCAN и цикла for создадим набор моделей для различных значений параметра эпсилон. В качестве параметра min_samples возьмём значение 2 умноженное на количество признаков. Во время цикла сохраним процент точек-выбросов.**

```python
from sklearn.cluster import DBSCAN
```

```python
outlier_percent = []
number_of_outliers = []

for eps in np.linspace(0.001,3,50):
    
    # Создаём модель
    dbscan = DBSCAN(eps=eps, min_samples=2*scaled_data.shape[1])
    dbscan.fit(scaled_data)
    
    # Сохраняем количество точек выбросов
    number_of_outliers.append(np.sum(dbscan.labels_ == -1))
    
    # Сохраняем процент точек-выбросов (количество выбросов как процент от общего количества точек)
    perc_outliers = 100 * np.sum(dbscan.labels_ == -1) / len(dbscan.labels_)
    
    outlier_percent.append(perc_outliers)
```

**Построим график lineplot для процента точек-выбросов в зависимости от значений эпсилон.**

```python
sns.lineplot(x=np.linspace(0.001,3,50),y=outlier_percent)
plt.ylabel("Percentage of Points Classified as Outliers")
plt.xlabel("Epsilon Value")
```

## DBSCAN для выбранного значения

**На основе графика, создадим модель DBSCAN с выбранным по колену графика значением эпсилон равное двум.**

```python
dbscan = DBSCAN(eps=2, min_samples=2*scaled_data.shape[1])
dbscan
```

**Построим график scatterplot для колонок Milk и Grocery, раскрашивая данные разными цветами по результатам работы модели DBSCAN.**

```python
    labels = dbscan.fit_predict(scaled_data)
    sns.scatterplot(data=df,x='Grocery',y='Milk',hue=labels,palette='Set1');
```

**Построим график scatterplot для колонок Milk и Detergents Paper, раскрашивая данные разными цветами по результатам работы модели DBSCAN.**

```python
sns.scatterplot(data=df,x='Detergents_Paper',y='Milk',hue=labels,palette='Set1');
```

**Создадим новую колонку в исходном датафрейме, назовём эту колонку "Labels" и запишем в неё результаты работы модели DBSCAN - значения labels.**

```python
df['Labels']=labels
```

```python
df.head()
```

**Вычислим средние значения расходов по различным категориям, причём отдельно для каждого из кластеров (включая кластер -1 для выбросов). Сравним расходы разных кластеров между собой (по отдельным категориям расходов).**

```python

df_grouped=df.drop(['Channel','Region'], axis=1).groupby('Labels').mean()
df_grouped
```

```python
plt.figure(figsize=(12,6), dpi=200)
sns.heatmap(df_grouped, annot=True)
```

```python
plt.figure(figsize=(12,6), dpi=200)
sns.heatmap(df_grouped.loc[{0,1}], annot=True)
```

**Выполним нормализацию таблицы с данными, полученной в предыдущем задании, с помощью MinMaxScaler - чтобы значения менялись в диапазоне от 0 до 1. После этого построим тепловую карту heatmap для получившихся значений.**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df_grouped)
scaled_df_grouped = pd.DataFrame(scaled_data,columns=df_grouped.columns,index=df_grouped.index)
scaled_df_grouped
```

```python
sns.heatmap(scaled_df_grouped, annot=True)
```

**Построим ещё одну тепловую карту - такую же, как и в предыдущем задании, но на этот раз предварительно удалим выбросы из рассмотрения.**

```python
sns.heatmap(scaled_df_grouped.drop(index=[-1]), annot=True)
```

**Глядя на тепловую карту, кластеры различаются наиболее заметно по колонке Detergents_Paper**
